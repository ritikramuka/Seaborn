```python
import seaborn as sns
print(sns.get_dataset_names())
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

**Matplotlib vs Seaborn**

```python
from matplotlib import pyplot as plt
import seaborn as sns
plt.scatter(df['sepal_length'],df['petal_length'])
plt.show()
```

```python
plt.scatter(df['sepal_length'],df['petal_length'])
sns.set()
plt.show()
```

```python
plt.scatter(df['sepal_length'],df['petal_length'])
sns.set_style("darkgrid", {'grid.color': '.5'})
sns.despine()
plt.show()
```

**COLOR PALETTE(Total 170)**

```python
sns.palplot(sns.color_palette("deep", 10))
```

Relational Plot

```python
sns.relplot(data = df,x='sepal_length',y='petal_length',hue="species")
```

```python
sns.relplot(data = df,x='sepal_length',y='petal_length',kind="line")
```

HISTOGRAM

```python
sns.distplot(df['petal_length'],kde = True)
```

Barplot

```python
sns.barplot(x = 'species', y = 'petal_length', data = df,ci=None )
plt.legend()
plt.show()
```

```python
sns.barplot(x = 'petal_length', y = 'species', data = df,orient = 'h',ci=None )
plt.legend()
plt.show()
```

Count Plot

```python
df2 = sns.load_dataset('titanic')
df2.head()
```

```python
sns.countplot(x = 'class', hue = 'who', data = df2, palette = 'magma')
plt.title('Survivors')
```

KDE Plot

```python
sns.kdeplot(df[df['species']=='setosa']['sepal_length'],df['sepal_width'],color='r', shade=True,  cmap="Reds") 
sns.kdeplot(df[df['species']=='virginica']['sepal_length'],df['sepal_width'],color='b', shade=True, cmap="Blues")
```

Boxplot

```python
info = df2[df2['fare']<=300]
```

```python
sns.boxplot(x="class", y="fare", data=info)
```

Heatmap

```python
flights=sns.load_dataset("flights")
flights.head()
```

```python
flights = flights.pivot("month", "year", "passengers")
sns.heatmap(flights)
```

Pair Plot

```python
sns.pairplot(df,hue = 'species',kind = "scatter")
```
